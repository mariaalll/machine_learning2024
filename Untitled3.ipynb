```python
# Данные
Nail_style = ["Шеллак", "Френч", "Обычный лак", "Гель-лак", "Акрил"]
Price = [2000, 1500, 1000, 3000, 3500]
Week = [4, 5, 4, 8, 6]

# 1. Среднее значение посещений салона
total_visits = sum(Week)
average_visits = total_visits / len(Week)
print("Среднее значение посещений салона:", average_visits)

# 2. Общее количество посещений салона
print("Общее количество посещений салона:", total_visits)

# 3. Выручка салона
total_revenue = 0
for i in range(len(Nail_style)):
    total_revenue += Price[i] * Week[i]
print("Выручка салона:", total_revenue)

# 4. Выручка по видам маникюра
print("Выручка по видам маникюра:")
for i in range(len(Nail_style)):
    revenue_per_style = Price[i] * Week[i]
    print(Nail_style[i], ":", revenue_per_style)

# 5. Средняя выручка в день по видам маникюра
print("Средняя выручка в день по видам маникюра:")
for i in range(len(Nail_style)):
    average_revenue_per_day = (Price[i] * Week[i]) / 7
    print(Nail_style[i], ":", average_revenue_per_day)
```
